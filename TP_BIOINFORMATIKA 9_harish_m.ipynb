*   Sp1: A:10%, T: 10%, G: 40%, C:40%

Hint:

*   Pelajari contoh penggunaan python function `def`, https://www.w3schools.com/python/python_functions.asp, atau dari sumber lain
*   Buat fungsinya, kemudian panggil fungsi tersebut untuk setiap kasus sp1, sp2, sp3.
<br>

**Output:** ditampilkan pie chart Sp1, Sp2, Sp3 sekaligus dari hasil pemanggilan fungsi

Jawab: <br> Fungsi pie chart:

````python
# Defining pie chart function
# percentage_sp = [A, T, G, C]
def pie_chart(percent, title):
  # Create a pie chart
  plt.figure(figsize=(2, 2))
  plt.pie(percent, labels = ['A', 'T', 'G', 'C'], autopct='%1.1f%%', startangle=140, colors=['lightcoral', 'lightblue', 'lightgreen', 'lightsalmon'])
  plt.title(title)
  plt.axis('equal')  # Equal aspect ratio ensures that the pie is drawn as a circle.
  # Display the pie chart
  plt.show()
````

Pemanggilan fungsi pie chart:

````python
percentage_sp1 = [10, 10, 40, 40]
percentage_sp2 = [25, 30, 15, 30]
percentage_sp3 = [10, 20, 35, 35]

pie_chart(percentage_sp1, 'Sp1')
pie_chart(percentage_sp2, 'Sp2')
pie_chart(percentage_sp3, 'Sp3')
````
